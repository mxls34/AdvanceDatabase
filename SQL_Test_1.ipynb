**2.4**

```python
# เขียนคำสั่ง SQL ข้อ 2.4 ตรงนี้
run("""
    SELECT name, total_spent, CASE
        WHEN total_spent < 1000 THEN 'ลูกค้าทั่วไป'
        WHEN total_spent < 10001 THEN 'ลูกค้าประจำ'
        WHEN total_spent > 10000 THEN 'ลูกค้า VIP'
        ELSE 'ไม่เคยซื้อ'
      END AS customer_tier
    FROM customer
""")
```
