## Part 2: Complex Conditions

**2.1** หาลูกค้าที่มีอายุอยู่ในช่วง 18-60 ปี (ใช้ `BETWEEN`)

````python
# TODO 2.1
con.execute("""
      SELECT * FROM customers WHERE age BETWEEN 18 AND 60
""").df()
````

````python
# ตัวอย่างผลลัพธ์ 2.1
````

**2.2** หาลูกค้าที่อยู่ใน "กรุงเทพฯ" หรือ "เชียงใหม่" หรือ "ภูเก็ต" เท่านั้น (ใช้ `IN`)

````python
# TODO 2.2
con.execute("""
        SELECT * FROM customers WHERE city IN ('กรุงเทพฯ', 'ภูเก็ต', 'เชียงใหม่')
""").df()
````

````python
# ตัวอย่างผลลัพธ์ 2.2
````

**2.3** หาแถวที่ `email` **ไม่มี** เครื่องหมาย `@` อยู่เลย (แสดงว่าอีเมลผิดรูปแบบ)
ใช้ `LIKE` ในการหา

> ระวังไม่ให้ NULL หรือค่าว่าง `''` มาปนในผลลัพธ์ด้วย ลองคิดดูว่าต้องเพิ่มเงื่อนไขอะไรอีก

````python
# TODO 2.3
con.execute("""
      SELECT * FROM customers WHERE email NOt LIKE '%@%'
""").df()
````

````python
# ตัวอย่างผลลัพธ์ 2.3
````

**2.4** ใช้ `CASE` แบ่งกลุ่มลูกค้าตามอายุเป็น 3 กลุ่ม: อายุน้อยกว่า 18 = 'เยาวชน',
18-59 = 'วัยทำงาน', 60 ปีขึ้นไป = 'ผู้สูงอายุ' ตั้งชื่อคอลัมน์ผลลัพธ์ว่า `age_group`

> ให้ลองสังเกตว่าถ้า `age` เป็น NULL ผลลัพธ์ของ CASE จะเป็นอย่างไร (ไปตกที่ ELSE หรือไม่?)

````python
# TODO 2.4
con.execute("""
      SELECT full_name, age,
        CASE
          WHEN age < 18 THEN 'เยาวชน'
          WHEN age < 60 THEN 'วัยทำงาน'
          ELSE 'ผู้สูงอายุ'
        END AS age_group
      FROM customers
""").df()
````

````python
# ตัวอย่างผลลัพธ์ 2.4
````
